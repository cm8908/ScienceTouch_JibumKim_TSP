```python
# 실행에 필요한 패키지를 불러옵니다 (pandas, numpy, matplotlib)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# weather_covid.csv 파일을 pandas를 이용해 불러옵니다.
data_df = pd.read_csv("weather_covid.csv", index_col=0)
data_df.head()  # 데이터프레임.head()를 이용해 데이터의 첫 다섯 줄을 확인할 수 있습니다.
```

```python
# 데이터프레임.info()를 이용해 각 열의 데이터 타입과 행의 개수를 알 수 있습니다.
data_df.info()
```

```python
# 데이터프레임.describe()를 통해 각 열의 통계량을 확인할 수 있습니다.
data_df.describe()
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # 한글 폰트가 깨지지 않게 출력되도록 설정
matplotlib.rcParams['axes.unicode_minus'] = False

# 데이터프레임.tail(150)을 통해 마지막 150개의 행을 가져와서 df1에 저장합니다.
df1 = data_df.tail(150)

x2 = df1.index  # 년월일(index)
y2 = df1.tail(150)["확진자수"]  # 확진자수
```

```python
# x축에 년월일, y축에 확진자수를 놓고 산점도를 그립니다.
plt.figure(figsize=(20, 5))
plt.scatter(x2, y2)
plt.xticks(np.arange(0, len(x2), 7))
plt.xticks(rotation=30)
plt.title("코로나19 일별 확진자수(인천)", fontsize=20)
plt.xlabel("년월일", fontsize=14)
plt.ylabel("확진자수", fontsize=14)
plt.show()
```

```python
# plt.hist를 이용하면 히스토그램을 그릴 수 있습니다.
plt.figure(figsize=(8, 5))
plt.hist(df1.확진자수, bins=10)
plt.xticks(np.arange(0, max(df1.확진자수), 500))
plt.title("코로나19 일별 확진자수(인천)", fontsize=20)
plt.xlabel("계급(확진자수)", fontsize=14)
plt.ylabel("빈도수(일수)", fontsize=14)
plt.show()
```
